-> operators are special symbols used to perform operations on variables and values.

->Python includes several types of operators, each serving different purposes.

-> Here's an overview of the main types of operators with examples:


                        **1. Arithmetic Operators := **

These operators perform basic arithmetic operations.

a = 5

result = (a == b)  # result is False

result = (a != b)  # result is True

-> Greater than (>):

result = (a > b)  # result is True

result = (a >= b)  # result is True

-> Less than or equal to (<=):

result = (a <= b)  # result is False

->Or (or):

result = (a > 2 or b > 5)  # result is True

->Not (not):

result = not (a > 2)  # result is False

These operators are used to assign values to variables.

                      **5. Bitwise Operators := **
These operators perform bit-level operations.

a = 5  # 0101 in binary
b = 3  # 0011 in binary

OR (|):

# Question : 5 explain the concept of type casting in Python with exampless



-> Type casting in Python refers to converting a value from one data type to another.

-> This is often necessary when performing operations that require a specific type or when you want to ensure that data is in the desired format.

-> Python provides several built-in functions for type casting.

-> Hereâ€™s a breakdown of how type casting works with examples:

#**1. Integer to Float**

When you need to convert an integer to a float, you can use the float() function.

# Integer
num = 5
# Convert to float
float_num = float(num)

print(float_num)  # Output: 5.0

#**2. Float to Integer**


Converting a float to an integer is done using the int() function. Note that this conversion truncates the decimal part, it doesnâ€™t round.

# Float
num = 5.7

# Convert to integer
int_num = int(num)

print(int_num)  # Output: 5

#**3. String to Integer**


To convert a string that represents an integer to an actual integer, use the int() function.

# String
str_num = "123"

# Convert to integer
int_num = int(str_num)

print(int_num)  # Output: 123


#**4. String to Float**

To convert a string that represents a floating-point number to a float, use the float() function.

# String
str_num = "123.45"

# Convert to float
float_num = float(str_num)

print(float_num)  # Output: 123.45

# **5. Integer to String**

You can convert an integer to a string using the str() function.

# Integer
num = 123

# Convert to string
str_num = str(num)

print(str_num)  # Output: '123'

#**6. Float to String **

To convert a float to a string, also use the str() function.

# Float
num = 123.45

# Convert to string
str_num = str(num)

print(str_num)  # Output: '123.45'

#**7. String to List or Tuple**

When working with strings, you might want to convert them into lists or tuples. For example, you can split a string into a list using the split() method and then cast it to a tuple if needed.

# String
str_data = "apple,banana,cherry"

# Convert to list
list_data = str_data.split(',')

print(list_data)  # Output: ['apple', 'banana', 'cherry']

# Convert list to tuple
tuple_data = tuple(list_data)

print(tuple_data)  # Output: ('apple', 'banana', 'cherry')

# **8. List to String**

To convert a list to a string, you can use the join() method.

# List
list_data = ['apple', 'banana', 'cherry']

# Convert to string
str_data = ','.join(list_data)

print(str_data)  # Output: 'apple,banana,cherry'

# 9. Complex Type Casting

Sometimes, you might need to perform more complex type conversions, such as converting a list of strings into a list of integers.

# List of strings
str_list = ['1', '2', '3']

# Convert to list of integers
int_list = [int(x) for x in str_list]

print(int_list)  # Output: [1, 2, 3]

Type casting is a crucial aspect of programming in Python, enabling you to ensure data is in the correct format for various operations. The primary functions used for type casting include:

list(), tuple(), set(): Converts to corresponding data structures.




# Question : 6   How do conditional statements work in Python? Illustrate with examples.


-> Conditional statements in Python allow you to execute different blocks of code based on certain conditions.

-> These statements enable you to control the flow of execution in your programs.

->The main types of conditional statements in Python are if, elif, and else.

-> Hereâ€™s a detailed look at how they work, with examples:

# Basic if Statement

The if statement evaluates a condition (an expression that returns True or False). If the condition is True, the block of code following the if statement is executed.

Example :

    x = 10
    if x > 5:
    print("x is greater than 5")

Output:

x is greater than 5

#if-else Statement

The if-else statement provides an alternative block of code to execute when the if condition is not met.

Example :

    x = 3
    if x > 5:
        print("x is greater than 5")
    else:
        print("x is not greater than 5")

Output:

x is not greater than 5

#if-elif-else Statement

The if-elif-else statement allows you to check multiple conditions. The elif (short for "else if") block provides additional conditions if the initial if condition is not met.

Example :
x = 7

    if x > 10:
        print("x is greater than 10")
    elif x > 5:
        print("x is greater than 5 but not greater than 10")
    else:
        print("x is 5 or less")
Output:

x is greater than 5 but not greater than 10

# Nested if Statements

You can nest if statements inside other if statements to handle more complex scenarios.


    x = 10
    y = 20

    if x > 5:
        if y > 15:
            print("x is greater than 5 and y is greater than 15")
        else:
            print("x is greater than 5 but y is not greater than 15")
    else:
        print("x is not greater than 5")
Output:

x is greater than 5 and y is greater than 15

# Combining Conditions

You can combine multiple conditions using logical operators like and, or, and not.


    x = 10
    y = 5

    if x > 0 and y > 0:
        print("Both x and y are positive")

    if not (x < 0):
        print("x is non-negative")
Output:

Both x and y are positive

x is non-negative

# 3. List Comprehensions

List comprehensions provide a concise way to create lists. They are often used to perform operations on each item in a sequence and build a new list.

Basic Example

    # Creating a list of squares using a list comprehension
    squares = [x**2 for x in range(5)]
    print(squares)
Output:

[0, 1, 4, 9, 16]

    # Creating a list of even numbers using a list comprehension
    even_numbers = [x for x in range(10) if x % 2 == 0]
    print(even_numbers)
Output:

[0, 2, 4, 6, 8]

    # Creating a 2D matrix using nested list comprehensions
    matrix = [[row * col for col in range(3)] for row in range(3)]
    print(matrix)

[[0, 0, 0],

[0, 1, 2],

 [0, 2, 4]]
