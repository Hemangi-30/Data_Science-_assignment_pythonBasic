## 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Generate synthetic data
num_samples = 100

# Generate features with some correlation
feature1 = np.random.normal(loc=0, scale=1, size=num_samples)
feature2 = feature1 * 0.5 + np.random.normal(loc=0, scale=1, size=num_samples)  # Correlated with feature1
feature3 = feature1 * -0.3 + feature2 * 0.7 + np.random.normal(loc=0, scale=1, size=num_samples)  # Correlated with feature1 and feature2

# Create a DataFrame
df = pd.DataFrame({
    'Feature1': feature1,
    'Feature2': feature2,
    'Feature3': feature3
})

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f', linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(20)

# Generate synthetic data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data for the heatmap
heatmap_data = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap
fig = px.imshow(heatmap_data,
                labels={'x': 'Month', 'y': 'Day', 'color': 'Average Sales'},
                title='Sales Heatmap Across Different Months and Days',
                color_continuous_scale='Viridis')

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# Generate synthetic data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the data for Plotly
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(
    z=z,
    x=x,
    y=y,
    colorscale='Viridis'
)])

# Update the layout
fig.update_layout(
    title='3D Surface Plot of $z = \sin(\sqrt{x^2 + y^2})$',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ),
    coloraxis_colorbar=dict(title='Z Value')
)

# Show the plot
fig.show()
```

## 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable Bokeh output in the notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 50  # Number of data points
x = np.random.uniform(0, 10, n)
y = np.random.uniform(0, 10, n)
sizes = np.random.uniform(10, 100, n)  # Marker sizes
colors = np.random.choice(['#f66', '#6f6', '#66f', '#ff6', '#6ff'], n)  # Marker colors

# Create a DataFrame
data = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

# Create a new plot with title and axis labels
p = figure(title="Bokeh Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add a scatter renderer with variable sizes and colors
p.scatter(x='x', y='y', size='sizes', color='colors', legend_field='colors', source=data)

# Update layout
p.legend.title = 'Colors'
p.legend.label_text_font_size = '8pt'

# Show the plot
show(p)
```

## 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.palettes import Category10
import pandas as pd

# Enable Bokeh output in the notebook
output_notebook()

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame
data = pd.DataFrame({
    'Fruit': fruits,
    'Count': counts
})

# Create a new plot with title and axis labels
p = figure(x_range=data['Fruit'], title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count',
           toolbar_location=None, tools="")

# Add a bar renderer with colors from Category10 palette
colors = Category10[len(fruits)]  # Get a list of colors

# Map each fruit to a color
data['color'] = colors

# Add a bar renderer
p.vbar(x='Fruit', top='Count', width=0.4, source=data, color='color', legend_field='Fruit')

# Update layout
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

## 4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable Bokeh output in the notebook
output_notebook()

# Generate data
np.random.seed(42)  # For reproducibility
data_hist = np.random.randn(1000)  # Random data
hist, edges = np.histogram(data_hist, bins=30)  # Compute histogram

# Create a DataFrame for Bokeh plotting
hist_data = pd.DataFrame({
    'left': edges[:-1],  # Use bin edges excluding the last one for left x values
    'right': edges[1:],  # Use bin edges excluding the first one for right x values
    'counts': hist
})

# Create a new plot with title and axis labels
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency',
           tools="", background_fill_color="#fafafa")

# Add a quad renderer
p.quad(top='counts', bottom=0, left='left', right='right', source=hist_data,
        fill_color='navy', line_color='white')

# Update layout
p.grid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None

# Show the plot
show(p)
```

## 5. Create a Bokeh heatmap using the provided dataset.

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Enable Bokeh output in the notebook
output_notebook()

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data and create a DataFrame
data = pd.DataFrame({
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
})

# Create a new plot with title and axis labels
p = figure(title="Heatmap of Random Data", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), toolbar_location=None, tools="")

# Create a color mapper
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data['value']), high=max(data['value']))

# Add a rect renderer to create the heatmap
p.rect(x='x', y='y', width=0.1, height=0.1, source=data, line_color=None, fill_color=mapper)

# Update layout
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
p.xaxis.major_label_orientation = "horizontal"
p.yaxis.major_label_orientation = "horizontal"

# Add a color bar
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
