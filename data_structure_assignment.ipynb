# question 1 : Discuss string slicing and provide examples.

--> String slicing in Python is a powerful feature that allows you to extract substrings from a larger string.

--> This is done using the slice notation, which is string[start:stop:step]. Here's a breakdown of how it works:

* Slice Notation
->start: The index where the slice begins (inclusive). If omitted, it defaults to the beginning of the string.

->stop: The index where the slice ends (exclusive). If omitted, it defaults to the end of the string.

-> step: The step between each index. If omitted, it defaults to 1.
##Examples

    text = "Hello, World!"

    # Slice from index 0 to 4 (exclusive)
    print(text[0:5])  # Output: 'Hello'

    # Slice from index 7 to the end
    print(text[7:])  # Output: 'World!'

    # Slice from the start to index 5 (exclusive)
    print(text[:6])  # Output: 'Hello,'
Using the step Parameter

    text = "Hello, World!"

    # Slice with step 2 from index 0 to 12 (exclusive)
    print(text[0:13:2])  # Output: 'Hlo ol!'

    # Slice with step -1 (reverse the string)
    print(text[::-1])  # Output: '!dlroW ,olleH'
Omitting Parameters

    text = "Hello, World!"

    # Slice the entire string (same as text[:])
    print(text[:])  # Output: 'Hello, World!'

    # Slice the entire string with a step of 3
    print(text[::3])  # Output: 'Hl r!'
Negative Indices

    text = "Hello, World!"

    # Slice with negative start index
    print(text[-6:-1])  # Output: 'World'

    # Slice with negative indices and a step
    print(text[-1:-6:-1])  # Output: 'World'

# question 2 :  Explain the key features of lists in Python.

--> Lists in Python are one of the most versatile and commonly used data structures.

->They are mutable, ordered collections that can hold items of different types.

->Here are the key features of Python lists:

## Key Features

1.Ordered:

-> Lists maintain the order of elements as they are added. Each element has a specific position (or index) starting from 0.

    my_list = ['apple', 'banana', 'cherry']
    print(my_list[1])  # Output: 'banana'

-> Lists are mutable, meaning you can change their content after they are created. You can modify, add, or remove elements.

    my_list = [1, 2, 3]

-> Lists can grow or shrink in size dynamically. You donâ€™t need to declare the size beforehand.

    my_list = [1, 2]

-> Lists can contain elements of different types, including other lists.

    my_list = [1, 'hello', 3.14, [1, 2, 3]]

-> You can access elements using indexing and extract parts of lists using slicing.

    my_list = ['a', 'b', 'c', 'd', 'e']
    print(my_list[1:4])  # Output: ['b', 'c', 'd']

6. Nested Lists:

-> Lists can contain other lists, allowing for complex data structures.

    nested_list = [[1, 2], [3, 4], [5, 6]]

# question 3 : Describe how to access, modify, and delete elements in a list with examples.

# 1. Accessing Elements

Using Indexing:

-> Lists are zero-indexed, meaning the first element is at index 0, the second at index 1, and so on. Negative indices count from the end of the list, with -1 being the last element.

    my_list = ['apple', 'banana', 'cherry']

    # Accessing elements using positive index
    print(my_list[0])  # Output: 'apple'

    # Accessing elements using negative index
    print(my_list[-1])  # Output: 'cherry'

    # Accessing a slice of the list
    print(my_list[1:3])  # Output: ['banana', 'cherry']


# 2. Modifying Elements

Using Indexing:

-> You can change the value of an element by accessing it via its index and assigning a new value.

    my_list = [10, 20, 30, 40]

    # Modifying an element
    my_list[2] = 35
    print(my_list)  # Output: [10, 20, 35, 40]

    # Modifying a slice of the list
    my_list[1:3] = [25, 30]
    print(my_list)  # Output: [10, 25, 30, 40]

Using List Methods:

-> Methods such as append(), extend(), insert(), and pop() can be used to modify the list.

    my_list = [1, 2, 3]

    # Appending an element
    my_list.append(4)
    print(my_list)  # Output: [1, 2, 3, 4]

    # Inserting an element at a specific position
    my_list.insert(1, 10)
    print(my_list)  # Output: [1, 10, 2, 3, 4]

    # Extending the list with another list
    my_list.extend([5, 6])
    print(my_list)  # Output: [1, 10, 2, 3, 4, 5, 6]

    # Removing an element by value
    my_list.remove(10)
    print(my_list)  # Output: [1, 2, 3, 4, 5, 6]

    # Removing an element by index
    removed_element = my_list.pop(3)
    print(my_list)  # Output: [1, 2, 3, 5, 6]
    print(removed_element)  # Output: 4

# 3. Deleting Elements

Using del,pop,remove  Statement:

-> You can use the del keyword to remove an element or a slice of elements.

    my_list = [10, 20, 30, 40]

    # Deleting an element by index
    del my_list[1]
    print(my_list)  # Output: [10, 30, 40]

    # Deleting a slice of elements
    del my_list[0:2]
    print(my_list)  # Output: [40]

    my_list = [1, 2, 3, 4]

    # Removing the last element
    last_element = my_list.pop()
    print(my_list)  # Output: [1, 2, 3]
    print(last_element)  # Output: 4

    # Removing an element by index
    element = my_list.pop(1)
    print(my_list)  # Output: [1, 3]
    print(element)  # Output: 2

    my_list = ['apple', 'banana', 'cherry', 'banana']

    # Removing an element by value
    my_list.remove('banana')
    print(my_list)  # Output: ['apple', 'cherry', 'banana']

# question 4 : Compare and contrast tuples and lists with examples.

-> Tuples and lists are both built-in data structures in Python that allow you to store collections of items. While they share some similarities, they have distinct differences. Here's a detailed comparison between tuples and lists, with examples:

## Similarities

1. Ordered Collections:

Both tuples and lists maintain the order of elements as they are inserted. You can access elements using indexing and slicing.

    my_list = [1, 2, 3]
    my_tuple = (1, 2, 3)

    print(my_list[1])  # Output: 2
    print(my_tuple[1])  # Output: 2
2. Support for Slicing:

Both can be sliced to extract parts of the collection.

    my_list = [10, 20, 30, 40]
    my_tuple = (10, 20, 30, 40)

    print(my_list[1:3])  # Output: [20, 30]
    print(my_tuple[1:3])  # Output: (20, 30)
3. Support for Iteration:

Both can be iterated over using loops.

    my_list = [1, 2, 3]
    my_tuple = (1, 2, 3)

    for item in my_list:
        print(item)  # Output: 1 2 3

    for item in my_tuple:
        print(item)  # Output: 1 2 3

        
## Differences

1. Mutability:

Lists --> are mutable, meaning their elements can be changed after creation. You can add, remove, or modify elements.

Tuples -->  are immutable, meaning once they are created, their elements cannot be changed.

    my_list = [1, 2, 3]
    my_list[1] = 20
    print(my_list)  # Output: [1, 20, 3]

    my_tuple = (1, 2, 3)
    # my_tuple[1] = 20  # This will raise a TypeError

2. Syntax:

Lists use square brackets [].

Tuples use parentheses ().

    my_list = [1, 2, 3]
    my_tuple = (1, 2, 3)

3. Performance:

-->Tuples generally have a smaller memory footprint and are faster than lists for certain operations because they are immutable. They are often used for fixed collections of items.

--> Lists are more flexible but might incur more overhead due to their mutability.

    import sys

    my_list = [1, 2, 3]
    my_tuple = (1, 2, 3)

    print(sys.getsizeof(my_list))  # Memory size of list
    print(sys.getsizeof(my_tuple))  # Memory size of tuple

4. Methods:

--> Lists come with various methods to modify the list, such as append(), extend(), insert(), remove(), pop(), sort(), and reverse().

--> Tuples have fewer methods, primarily count() and index(), due to their immutability.

    my_list = [1, 2, 3]
    my_list.append(4)
    my_list.sort()
    print(my_list)  # Output: [1, 2, 3, 4]

    my_tuple = (1, 2, 3)
    print(my_tuple.count(2))  # Output: 1
    print(my_tuple.index(3))  # Output: 2

5. Use Cases:

--> Lists are used when you need a collection that may change during the program execution, such as a list of items in a shopping cart or a list of users.

-->Tuples are used for fixed collections that should not change, such as coordinates, fixed configuration values, or function return values that should be grouped together.

# question 5  : Describe the key features of sets and provide examples of their use.


## Key Features of Sets

1. Unordered Collection:

Sets do not maintain any order for their elements. The order of elements is not guaranteed, and the elements are not indexed.

    my_set = {1, 2, 3, 4}
    print(my_set)  # Output might be {1, 2, 3, 4} but the order is not guaranteed
2. Unique Elements:

Sets automatically enforce uniqueness; duplicate elements are not allowed. Adding a duplicate element will not change the set.

    my_set = {1, 2, 2, 3}
    print(my_set)  # Output: {1, 2, 3}
3. Mutable:

Sets are mutable, meaning you can add or remove elements. However, the elements themselves must be immutable types (e.g., integers, strings, tuples).

    my_set = {1, 2, 3}
    my_set.add(4)  # Add an element
    my_set.remove(2)  # Remove an element
    print(my_set)  # Output: {1, 3, 4}
4. No Indexing:

Unlike lists or tuples, sets do not support indexing or slicing. You cannot access elements by position.

    my_set = {1, 2, 3}
    # my_set[0]  # This will raise a TypeError
5. Set Operations:

Sets support standard mathematical set operations like union, intersection, difference, and symmetric difference.

    set1 = {1, 2, 3}
    set2 = {3, 4, 5}

    # Union
    union_set = set1 | set2
    print(union_set)  # Output: {1, 2, 3, 4, 5}

    # Intersection
    intersection_set = set1 & set2
    print(intersection_set)  # Output: {3}

    # Difference
    difference_set = set1 - set2
    print(difference_set)  # Output: {1, 2}

    # Symmetric Difference
    sym_diff_set = set1 ^ set2
    print(sym_diff_set)  # Output: {1, 2, 4, 5}
6. Membership Testing:

Testing whether an element is in a set is efficient and can be done using the in keyword.

    my_set = {1, 2, 3}
    print(2 in my_set)  # Output: True
    print(4 in my_set)  # Output: False
7. Set Methods:

Sets come with various methods for manipulation, including add(), discard(), remove(), pop(), clear(), and copy().

    my_set = {1, 2, 3}

    # Add an element
    my_set.add(4)

    # Discard an element (does not raise an error if the element is not present)
    my_set.discard(2)

    # Remove an element (raises KeyError if the element is not present)
    my_set.remove(1)

    # Remove and return an arbitrary element
    popped_element = my_set.pop()

    # Clear all elements
    my_set.clear()

    # Copy the set
    copied_set = my_set.copy()
8. Frozensets:

Python also provides frozenset, which is an immutable version of a set. It supports set operations but does not allow modification.

    my_frozenset = frozenset([1, 2, 3])

    # frozenset operations
    print(1 in my_frozenset)  # Output: True
    # my_frozenset.add(4)  # This will raise an AttributeError since frozensets are immutable

## Use Cases for Sets
1. Removing Duplicates:

Sets are useful for eliminating duplicate elements from a collection, as they inherently store only unique items.

    my_list = [1, 2, 2, 3, 4, 4, 5]
    unique_elements = set(my_list)
    print(unique_elements)  # Output: {1, 2, 3, 4, 5}
2. Membership Testing:

Sets provide efficient O(1) average time complexity for membership testing, making them suitable for scenarios where quick lookups are necessary.

    valid_users = {'Alice', 'Bob', 'Charlie'}
    user_to_check = 'Bob'

    if user_to_check in valid_users:
        print(f'{user_to_check} is a valid user.')
3. Set Operations:

Sets support mathematical set operations such as union, intersection, difference, and symmetric difference, making them ideal for operations involving multiple collections.

    data_a = {1, 2, 3, 4, 5}
    data_b = {4, 5, 6, 7, 8}

    # Union
    all_unique = data_a | data_b
    print(all_unique)  # Output: {1, 2, 3, 4, 5, 6, 7, 8}

    # Intersection
    common_elements = data_a & data_b
    print(common_elements)  # Output: {4, 5}

    # Difference
    unique_to_a = data_a - data_b
    print(unique_to_a)  # Output: {1, 2, 3}

    # Symmetric Difference
    sym_diff_set = data_a ^ data_b
    print(sym_diff_set)  # Output: {1, 2, 3, 6, 7, 8}
4. Efficient Data Structures:

Sets are useful for implementing data structures such as hash sets and bloom filters, where the primary operations involve checking for membership or maintaining unique elements.

    seen_items = set()
    for item in items:
        if item not in seen_items:
            # Process the item
            seen_items.add(item)
5. Set Membership and Relationships:

Sets are ideal for problems involving group memberships or relationships, such as finding common elements between two groups or tracking unique elements in a dataset.

    students_a = {'John', 'Paul', 'George'}
    students_b = {'Paul', 'Ringo'}

    # Find students who are in both groups
    common_students = students_a & students_b
    print(common_students)  # Output: {'Paul'}
6. Cache Implementation:

Sets can be used to implement caches or memoization techniques, where you need to keep track of previously seen items or results.

    cache = set()

    def compute(value):
        if value in cache:
            return 'Cached Result'
        # Expensive computation
        result = value * value
        cache.add(value)
        return result

# question 7 :  Describe how to add,modify, and delete items on a dictionary with examples.

## 1. Adding Items
You can add items to a dictionary by assigning a value to a new key. If the key already exists, the value will be updated; if the key does not exist, a new key-value pair will be added.

Example:

    # Initial dictionary
    my_dict = {'name': 'Alice', 'age': 25}

    # Adding a new key-value pair
    my_dict['city'] = 'New York'
    print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}

    # Adding another new key-value pair
    my_dict.update({'email': 'alice@example.com'})
    print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York', 'email': 'alice@example.com'}
## 2. Modifying Items

To modify an existing item in a dictionary, you simply assign a new value to the existing key.

Example:

    # Initial dictionary
    my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}

    # Modifying an existing key-value pair
    my_dict['age'] = 26
    print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'New York'}

    # Modifying another existing key-value pair
    my_dict['city'] = 'Los Angeles'
    print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'Los Angeles'}

## 3. Deleting Items

You can delete items from a dictionary using several methods:

Using the del statement: Removes the key-value pair with the specified key.
Using the pop() method: Removes the key-value pair with the specified key and returns the value.
Using the popitem() method: Removes and returns an arbitrary key-value pair.
Using the clear() method: Removes all items from the dictionary.
Examples:

    # Initial dictionary
    my_dict = {'name': 'Alice', 'age': 26, 'city': 'Los Angeles'}

    # Using the `del` statement
    del my_dict['city']
    print(my_dict)  # Output: {'name': 'Alice', 'age': 26}

    # Using the `pop()` method
    age = my_dict.pop('age')
    print(my_dict)  # Output: {'name': 'Alice'}
    print(age)      # Output: 26

    # Using the `popitem()` method
    item = my_dict.popitem()
    print(my_dict)  # Output: {}
    print(item)     # Output: ('name', 'Alice')

    # Using the `clear()` method
    my_dict.clear()
    print(my_dict)  # Output: {}

# Question 8 :  Discuss the importance of dictionary keys being immutable and provide examples

## Importance of Immutable Keys
1. Hash Consistency:

Dictionaries in Python use hashing to quickly access values associated with keys. The hash value of a key is computed based on its contents. If a key were mutable (i.e., its contents could change), the hash value might change after the key is added to the dictionary. This would disrupt the dictionary's internal organization and could lead to unpredictable behavior or errors.

2. Data Integrity:

Immutable keys ensure that once a key is added to a dictionary, its identity and hash value remain unchanged. This guarantees that the key-value pair remains consistent and retrievable throughout the life of the dictionary.

3. Performance:

Hash-based lookups in dictionaries are very efficient. If dictionary keys were mutable, the dictionary would need to constantly check and rehash keys, leading to performance degradation. Immutable keys help maintain fast lookups and overall dictionary performance.

## Ensuring Immutability of Keys
To ensure that the keys you use in a dictionary are immutable, follow these guidelines:

Use immutable data types: Strings, numbers, and tuples (with immutable elements) are safe choices for keys.
Avoid using mutable types: Lists, dictionaries, and sets should not be used as keys.
## Example: Using Immutable Keys in a Real-World Scenario
Consider a scenario where you need to map coordinates to city names. Since tuples (which can store coordinates) are immutable, they are suitable as keys for this mapping.

    # Mapping coordinates to city names
    city_locations = {
        (40.7128, -74.0060): 'New York',
        (34.0522, -118.2437): 'Los Angeles',
        (51.5074, -0.1278): 'London'
    }

    # Retrieving city name by coordinates
    location = (34.0522, -118.2437)
    city = city_locations.get(location, 'Unknown')
    print(city)  # Output: Los Angeles
