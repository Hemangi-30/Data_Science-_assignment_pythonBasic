# 1. What is hypothesis testing in statistics?
   Hypothesis testing is a method used to make decisions or inferences about population parameters based on sample data. It involves testing an assumption (hypothesis) about a population.

# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

   * **Null hypothesis (H₀):** The default assumption that there is no effect or difference.
   * **Alternative hypothesis (H₁ or Ha):** The claim that there is an effect or difference.
     The test aims to determine whether there is enough evidence to reject H₀ in favor of H₁.

# 3. What is the significance level in hypothesis testing, and why is it important?

   The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 or 0.01. It defines the threshold for statistical significance.

# 4. What does a P-value represent in hypothesis testing?
   The P-value is the probability of observing the sample results (or more extreme) assuming the null hypothesis is true.

# 5. How do you interpret the P-value in hypothesis testing?

   * If **P ≤ α**, reject the null hypothesis (significant result).
   * If **P > α**, fail to reject the null hypothesis (not significant).

# 6. What are Type I and Type II errors in hypothesis testing?

   * **Type I error (α):** Rejecting a true null hypothesis.
   * **Type II error (β):** Failing to reject a false null hypothesis.

#7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

   * **One-tailed test:** Tests for a difference in one direction (e.g., greater than).
   * **Two-tailed test:** Tests for a difference in both directions (e.g., not equal to).

#8. What is the Z-test, and when is it used in hypothesis testing?
   A **Z-test** is used to test population means when the population standard deviation is known and the sample size is large (n ≥ 30).

#9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

   $$
   Z = \frac{\bar{x} - \mu}{\sigma/\sqrt{n}}
   $$

   It shows how many standard deviations the sample mean ($\bar{x}$) is from the population mean ($\mu$) under the null hypothesis.

# 10. What is the T-distribution, and when should it be used instead of the normal distribution?

   The **T-distribution** is used when the population standard deviation is unknown and the sample size is small (n < 30). It accounts for extra variability.

#11. What is the difference between a Z-test and a T-test?

* **Z-test:** Known population standard deviation, large sample.
* **T-test:** Unknown population standard deviation, small sample.

#12. What is the T-test, and how is it used in hypothesis testing?
  A **T-test** compares the means of groups or a sample mean to a known value to determine if differences are statistically significant.

#13. What is the relationship between Z-test and T-test in hypothesis testing?
   Both test hypotheses about means; the T-test is used when the standard deviation is unknown or sample size is small, while the Z-test is used with known standard deviation and large sample size.

#14. What is a confidence interval, and how is it used to interpret statistical results?
  A **confidence interval (CI)** is a range of values within which the population parameter is expected to lie, with a certain level of confidence (e.g., 95%).

#15. What is the margin of error, and how does it affect the confidence interval?
   The **margin of error** determines the width of the confidence interval. A larger margin means a wider interval, indicating more uncertainty.

#16. How is Bayes' Theorem used in statistics, and what is its significance?
   **Bayes’ Theorem** updates the probability of a hypothesis based on new evidence. It is fundamental to Bayesian inference and decision-making under uncertainty.

#17. What is the Chi-square distribution, and when is it used?
  The **Chi-square distribution** is used for tests involving categorical data, such as tests for independence or goodness-of-fit.

#18. What is the Chi-square goodness of fit test, and how is it applied?
   This test determines if observed categorical data fit a specified distribution. It compares observed vs. expected frequencies.

#19. What is the F-distribution, and when is it used in hypothesis testing?

  The F-distribution is used in comparing variances and in ANOVA tests. It helps test if group variances are significantly different.

#20. What is an ANOVA test, and what are its assumptions?

  ANOVA (Analysis of Variance)** tests whether there are statistically significant differences between the means of three or more groups.
  Assumptions:

* Independence of observations
* Normally distributed groups
* Equal variances across groups (homogeneity of variance)

# 22. What are the different types of ANOVA tests?

* **One-way ANOVA:** One independent variable
* **Two-way ANOVA:** Two independent variables
* **Repeated measures ANOVA:** Same subjects under different conditions

# 23. What is the F-test, and how does it relate to hypothesis testing?

  The **F-test** compares two variances or multiple group means. It is used in ANOVA and regression analysis to test overall significance.

#1.  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results ?

````python
from scipy.stats import norm
import numpy as np

# Known parameters
population_mean = 100
population_std = 15
sample = np.array([102, 98, 101, 97, 100, 99, 105, 95, 98, 103])
sample_size = len(sample)
sample_mean = np.mean(sample)

# Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
````

#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean?

````python
def one_sample_z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p = 2 * (1 - norm.cdf(abs(z)))
    return z, p

sample_data = np.random.normal(102, 10, 40)
z, p = one_sample_z_test(sample_data, 100, 10)
print(f"Z: {z:.4f}, P-value: {p:.4f}")
````

# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    mu0: mean under null hypothesis
    mu1: mean under alternative hypothesis
    sigma: population standard deviation
    n: sample size
    alpha: significance level (Type I error)
    """
    se = sigma / np.sqrt(n)
    z_crit = norm.ppf(1 - alpha)  # one-tailed test
    x_crit = mu0 + z_crit * se    # critical value in original units

    # Define x range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions under H0 and H1
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    # Plot distributions
    plt.figure(figsize=(10, 5))
    plt.plot(x, y0, label='H₀ Distribution', color='blue')
    plt.plot(x, y1, label='H₁ Distribution', color='orange')

    # Shade Type I error region (alpha)
    x_alpha = np.linspace(x_crit, x.max(), 1000)
    plt.fill_between(x_alpha, norm.pdf(x_alpha, mu0, se), color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II error region (beta)
    x_beta = np.linspace(x.min(), x_crit, 1000)
    plt.fill_between(x_beta, norm.pdf(x_beta, mu1, se), color='green', alpha=0.3, label='Type II Error (β)')

    # Vertical line at critical value
    plt.axvline(x_crit, color='black', linestyle='--', label='Critical Value')

    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Compute Type II error (β)
    beta = norm.cdf(x_crit, mu1, se)
    power = 1 - beta

    print(f"Critical Value (x_crit): {x_crit:.2f}")
    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Statistical Power: {power:.4f}")
````

# 9 . Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

````python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean using the T-distribution.

    Parameters:
        sample (array-like): List or array of sample data
        confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
        (lower_bound, upper_bound): Tuple containing the confidence interval
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return round(lower_bound, 2), round(upper_bound, 2)
data = [102, 100, 98, 95, 101, 104, 97, 99, 103, 96]
ci = confidence_interval(data, confidence=0.95)
print(f"95% Confidence Interval: {ci}")
````

# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

````python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for the sample mean.

    Parameters:
        sample (array-like): List or array of sample data
        confidence (float): Confidence level (default is 0.95 for 95%)

    Returns:
        margin_of_error (float)
    """
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # Standard error of the mean
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)  # Critical t-value

    moe = t_crit * std_err
    return round(moe, 3)
data = [50, 52, 49, 51, 53, 54, 48, 50, 52, 49]
confidence_level = 0.95

moe = margin_of_error(data, confidence_level)
print(f"Margin of Error (at {int(confidence_level * 100)}% confidence): ±{moe}")
````

# 11 .Implement a Bayesian inference method using Bayes' Theorem in Python and explain the processD



##  **Bayes’ Theorem Recap**

Bayes’ Theorem allows you to **update your beliefs** about the probability of a hypothesis $H$ after observing new evidence $E$:

$$
P(H|E) = \frac{P(E|H) \cdot P(H)}{P(E)}
$$

Where:

* $P(H)$ = prior probability (initial belief about H)
* $P(E|H)$ = likelihood (how likely E is given H)
* $P(E)$ = marginal probability (probability of evidence)
* $P(H|E)$ = posterior probability (updated belief)

---

## **Python Implementation**

```python
def bayes_theorem(prior_H, likelihood_E_given_H, prior_not_H, likelihood_E_given_not_H):
    """
    Applies Bayes' Theorem to compute P(H|E)
    
    Parameters:
        prior_H: P(H) - Prior probability of hypothesis
        likelihood_E_given_H: P(E|H) - Likelihood of evidence given hypothesis
        prior_not_H: P(~H) - Prior probability of alternative hypothesis
        likelihood_E_given_not_H: P(E|~H) - Likelihood of evidence given alternative

    Returns:
        posterior_H: P(H|E)
    """
    numerator = likelihood_E_given_H * prior_H
    denominator = numerator + (likelihood_E_given_not_H * prior_not_H)
    posterior_H = numerator / denominator
    return round(posterior_H, 4)
```

---

##  **Example Use Case**

Let’s say:

* A disease affects **1%** of the population → $P(Disease) = 0.01$
* The test is **95% accurate** if the person **has** the disease → $P(Positive|Disease) = 0.95$
* The test gives **false positives** 5% of the time → $P(Positive|No Disease) = 0.05$

Now, calculate the probability that someone **actually has the disease** if they test **positive**.

```python
# Inputs
P_Disease = 0.01
P_Positive_given_Disease = 0.95
P_NoDisease = 1 - P_Disease
P_Positive_given_NoDisease = 0.05

# Apply Bayes' Theorem
posterior = bayes_theorem(P_Disease, P_Positive_given_Disease, P_NoDisease, P_Positive_given_NoDisease)

print(f"Probability of Disease given Positive Test (Posterior): {posterior:.4f}")
```

---

###  **Explanation of the Process**

* **Prior**: Start with a 1% belief that someone has the disease.
* **Likelihood**: The test is accurate 95% of the time.
* **Update**: After a positive test result, Bayes’ Theorem updates your belief.
* **Posterior**: The final probability reflects both the test result and the rarity of the disease.

 In this example, the **posterior is around 16%**, not 95%! This shows how rare conditions can have **low posterior probabilities**, even after a positive test.

# 12 .  Perform a Chi-square test for independence between two categorical variables in Python.

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Example: Create a contingency table
# Suppose we surveyed 100 people about preference for a product by gender

# Rows: Gender (Male, Female)
# Columns: Preference (Like, Dislike)
data = np.array([[30, 10],   # Male: 30 like, 10 dislike
                 [20, 40]])  # Female: 20 like, 40 dislike

# Convert to DataFrame for clarity (optional)
df = pd.DataFrame(data, index=['Male', 'Female'], columns=['Like', 'Dislike'])
print("Contingency Table:")
print(df)

# Perform Chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(data)

print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(pd.DataFrame(np.round(expected, 2), index=['Male', 'Female'], columns=['Like', 'Dislike']))

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("\nResult: Reject the null hypothesis.")
    print("Interpretation: There is a significant association between gender and preference.")
else:
    print("\nResult: Fail to reject the null hypothesis.")
    print("Interpretation: No significant association between gender and preference.")
````

# 13 . Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

````python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for Chi-square test from observed frequencies.

    Parameters:
        observed (2D array-like): Observed frequency contingency table

    Returns:
        expected (2D np.array): Expected frequencies table
    """
    observed = np.array(observed)
    row_sums = observed.sum(axis=1, keepdims=True)
    col_sums = observed.sum(axis=0, keepdims=True)
    total = observed.sum()

    expected = row_sums @ col_sums / total
    return expected

# Example observed data (contingency table)
observed_data = [[30, 10],
                 [20, 40]]

expected_freq = calculate_expected_frequencies(observed_data)

# Display results
df_expected = pd.DataFrame(np.round(expected_freq, 2),
                           index=['Row1', 'Row2'],
                           columns=['Col1', 'Col2'])
print("Expected Frequencies:")
print(df_expected)
````

# 14 . Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

````python
import numpy as np
from scipy.stats import chisquare

observed = np.array([50, 30, 20])
expected = np.array([33.33, 33.33, 33.33])

# Scale expected frequencies to match sum of observed frequencies exactly
expected = expected * observed.sum() / expected.sum()

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Interpretation: The observed distribution differs significantly from the expected distribution.")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Interpretation: The observed distribution is consistent with the expected distribution.")
````

# 15 . Create a Python script to simulate and visualize the Chi-square distribution and discuss its characteristics .

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

def plot_chi_square(dfs, sample_size=1000):
    """
    Simulate and plot Chi-square distributions for different degrees of freedom.

    Parameters:
        dfs (list): List of degrees of freedom to simulate
        sample_size (int): Number of samples per distribution
    """
    plt.figure(figsize=(10, 6))

    x = np.linspace(0, 30, 500)

    for df in dfs:
        # Theoretical PDF
        pdf = chi2.pdf(x, df)

        # Simulated samples
        samples = np.random.chisquare(df, size=sample_size)

        # Plot histogram of samples
        plt.hist(samples, bins=30, density=True, alpha=0.4, label=f'Simulated (df={df})')

        # Plot theoretical PDF
        plt.plot(x, pdf, lw=2, label=f'Theoretical PDF (df={df})')

    plt.title('Chi-square Distribution: Simulated vs Theoretical')
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Degrees of freedom to visualize
degrees_of_freedom = [1, 2, 5, 10]

plot_chi_square(degrees_of_freedom)
````

# 16 . Implement an F-test using Python to compare the variances of two random samples.

````python
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare variances of two samples.

    Parameters:
        sample1, sample2: arrays or lists of sample data
        alpha: significance level (default 0.05)

    Returns:
        f_stat: F statistic value
        p_value: two-tailed p-value
        conclusion: string interpreting the test
    """
    var1 = np.var(sample1, ddof=1)  # sample variance with Bessel's correction
    var2 = np.var(sample2, ddof=1)

    # Make sure F >= 1 by putting the larger variance in numerator
    if var1 > var2:
        f_stat = var1 / var2
        dfn = len(sample1) - 1  # degrees of freedom numerator
        dfd = len(sample2) - 1  # degrees of freedom denominator
    else:
        f_stat = var2 / var1
        dfn = len(sample2) - 1
        dfd = len(sample1) - 1

    # Two-tailed p-value for the F-test
    p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

    if p_value < alpha:
        conclusion = "Reject null hypothesis: Variances are significantly different."
    else:
        conclusion = "Fail to reject null hypothesis: No significant difference in variances."

    return f_stat, p_value, conclusion


# Example usage with random samples
np.random.seed(0)
sample1 = np.random.normal(loc=0, scale=1, size=30)
sample2 = np.random.normal(loc=0, scale=1.5, size=30)

f_stat, p_value, conclusion = f_test(sample1, sample2)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(conclusion)
````

# 24 .  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results .

````python
import numpy as np
from scipy.stats import chi2

# Sample data
data = np.array([10, 12, 9, 11, 13, 8, 14, 10, 12, 11])

# Hypothesized population variance
sigma0_squared = 4  # example value

# Sample size and variance
n = len(data)
sample_variance = np.var(data, ddof=1)

# Calculate test statistic
chi2_stat = (n - 1) * sample_variance / sigma0_squared

# Degrees of freedom
df = n - 1

# Significance level
alpha = 0.05

# Calculate critical values for two-tailed test
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)

# Calculate p-value for two-tailed test
if chi2_stat < (n - 1):
    p_value = 2 * chi2.cdf(chi2_stat, df)
else:
    p_value = 2 * (1 - chi2.cdf(chi2_stat, df))

# Print results
print(f"Sample variance: {sample_variance:.4f}")
print(f"Chi-square statistic: {chi2_stat:.4f}")
print(f"Degrees of freedom: {df}")
print(f"Critical values: {chi2_lower:.4f}, {chi2_upper:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if chi2_stat < chi2_lower or chi2_stat > chi2_upper:
    print("Reject the null hypothesis: population variance is significantly different from hypothesized variance.")
else:
    print("Fail to reject the null hypothesis: insufficient evidence to say population variance differs from hypothesized variance.")
````

# 26 . Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results .

````python
import matplotlib.pyplot as plt
import numpy as np
def f_test_and_visualize(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare variances of two samples, interpret, and visualize.

    Parameters:
        sample1, sample2: arrays or lists of sample data
        alpha: significance level (default 0.05)

    Returns:
        f_stat: F statistic value
        p_value: two-tailed p-value
        conclusion: string interpreting the test
    """
    var1 = np.var(sample1, ddof=1)  # sample variance with Bessel's correction
    var2 = np.var(sample2, ddof=1)

    # Ensure F >= 1 by putting the larger variance in numerator
    if var1 > var2:
        f_stat = var1 / var2
        dfn = len(sample1) - 1  # degrees of freedom numerator
        dfd = len(sample2) - 1  # degrees of freedom denominator
        larger_var_label = 'Sample 1'
        smaller_var_label = 'Sample 2'
    else:
        f_stat = var2 / var1
        dfn = len(sample2) - 1
        dfd = len(sample1) - 1
        larger_var_label = 'Sample 2'
        smaller_var_label = 'Sample 1'

    # Two-tailed p-value for the F-test
    # The p-value is the probability of observing an F-statistic as extreme as,
    # or more extreme than, the one calculated, under the null hypothesis.
    # For a two-tailed test, we consider the probability of the F-statistic
    # being greater than the observed F_stat or less than 1/F_stat.
    # Since F is always >= 1 in our calculation (by putting the larger variance in the numerator),
    # we only need to consider the upper tail, and then double the p-value.
    p_value = 2 * (1 - f.cdf(f_stat, dfn, dfd))

    # Interpret the result
    if p_value < alpha:
        conclusion = "Reject null hypothesis: Variances are significantly different."
    else:
        conclusion = "Fail to reject null hypothesis: No significant difference in variances."

    # --- Visualization ---
    plt.figure(figsize=(12, 6))

    # Plot 1: Comparison of Sample Variances (Bar plot or text)
    plt.subplot(1, 2, 1)
    sns.barplot(x=['Sample 1', 'Sample 2'], y=[var1, var2])
    plt.title('Sample Variances')
    plt.ylabel('Variance')
    plt.text(0, var1, f'{var1:.2f}', ha='center', va='bottom')
    plt.text(1, var2, f'{var2:.2f}', ha='center', va='bottom')


    # Plot 2: F-distribution with Calculated F-statistic and Critical Region
    plt.subplot(1, 2, 2)
    # Define x range for F-distribution PDF
    x = np.linspace(f.ppf(0.001, dfn, dfd), f.ppf(0.999, dfn, dfd), 1000) # Plot where PDF is non-zero

    # Plot the F-distribution PDF
    plt.plot(x, f.pdf(x, dfn, dfd), label=f'F-distribution (dfn={dfn}, dfd={dfd})')

    # Add vertical line for calculated F-statistic
    plt.axvline(f_stat, color='red', linestyle='--', label=f'Calculated F-stat = {f_stat:.2f}')

    # Add vertical line(s) for critical value(s) for two-tailed test
    # Since we put the larger variance in the numerator, we compare F_stat to the upper critical value at alpha/2
    f_critical_upper = f.ppf(1 - alpha/2, dfn, dfd)
    plt.axvline(f_critical_upper, color='green', linestyle=':', label=f'Upper Critical F ({1-alpha/2:.3f}) = {f_critical_upper:.2f}')

    # Shade the rejection region(s)
    # Upper tail
    x_rejection_upper = np.linspace(f_critical_upper, x.max(), 1000)
    plt.fill_between(x_rejection_upper, f.pdf(x_rejection_upper, dfn, dfd), color='orange', alpha=0.3, label='Rejection Region (alpha/2)')


    plt.title('F-Distribution and F-Test')
    plt.xlabel('F-statistic')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True, linestyle='--')
    plt.xlim(0, plt.xlim()[1]) # Ensure x-axis starts at 0
    plt.tight_layout()
    plt.show()


    # --- Interpretation ---
    print("\n--- F-Test Results ---")
    print(f"Sample 1 Variance: {var1:.4f} (n={len(sample1)})")
    print(f"Sample 2 Variance: {var2:.4f} (n={len(sample2)})")
    print(f"F-statistic ({larger_var_label} Var / {smaller_var_label} Var): {f_stat:.4f}")
    print(f"Numerator Degrees of Freedom (dfn): {dfn}")
    print(f"Denominator Degrees of Freedom (dfd): {dfd}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Critical F-value (Upper Tail, alpha/2): {f_critical_upper:.4f}")
    print(f"P-value (Two-tailed): {p_value:.4f}")
    print(f"\nConclusion: {conclusion}")
    if p_value < alpha:
        print(f"  Since P-value ({p_value:.4f}) < alpha ({alpha}), we have sufficient evidence to reject the null hypothesis.")
    else:
        print(f"  Since P-value ({p_value:.4f}) > alpha ({alpha}), we do not have sufficient evidence to reject the null hypothesis.")
    print("-" * 20)


# Example usage with random samples
np.random.seed(42)
sample1 = np.random.normal(loc=10, scale=5, size=25) # Mean 10, Std Dev 5 (Variance 25)
sample2 = np.random.normal(loc=12, scale=8, size=30) # Mean 12, Std Dev 8 (Variance 64)

f_test_and_visualize(sample1, sample2)

# Example with samples from populations with equal variances
np.random.seed(43)
sample3 = np.random.normal(loc=50, scale=10, size=28) # Mean 50, Std Dev 10 (Variance 100)
sample4 = np.random.normal(loc=55, scale=10, size=35) # Mean 55, Std Dev 10 (Variance 100)

print("\nTesting samples from populations with equal variances:")
f_test_and_visualize(sample3, sample4)
````
