#Quetion 3 : Explain the following terms: 1.The difference in np.array, np.asarray and np.asanyarray.  2. The difference between Deep copy and shallow copy.


1. np.array, np.asarray, and np.asanyarray
np.array:

Creates a new array from any object exposing the array interface (like lists or tuples).
Always returns a new instance.
Can take an optional dtype parameter to specify the data type.
np.asarray:

Converts input to an array but does not copy the data if the input is already an array of the same type.
Useful for avoiding unnecessary copies to save memory.
np.asanyarray:

Similar to np.asarray, but it allows for subclasses of ndarray (like masked arrays or matrix).
Returns an array of the same type if the input is already an array subclass.
2. Deep Copy vs. Shallow Copy
Shallow Copy:

Creates a new object, but inserts references into it to the objects found in the original.
Changes made to mutable objects in the shallow copy will affect the original (and vice versa).
Deep Copy:

Creates a new object and recursively copies all objects found in the original.
Changes made to objects in the deep copy do not affect the original.
In summary, shallow copies reference the same objects, while deep copies create completely independent copies.

#Quetion 5 : Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array perform the following operations:a) Extract all even integers from array. b) Extract all odd integers from array.

#Quetion 14 : Create a time-series dataset in Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib: a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity'). b) Label the x-axis as 'Date'. c) Set the title of the plot as 'Temperature and Humidity Over Time'.


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Create random temperature and humidity data
temperature = np.random.uniform(low=15, high=30, size=100)  # Random temperatures between 15C and 30C
humidity = np.random.uniform(low=30, high=90, size=100)     # Random humidity levels between 30% and 90%

# Create a DataFrame
data = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Set the 'Date' column as the index
data.set_index('Date', inplace=True)

# a) Plot 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature
ax1.plot(data.index, data['Temperature'], color='red', label='Temperature', marker='o')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(data.index, data['Humidity'], color='blue', label='Humidity', marker='x')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# b) Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
````

#Quetion 15 : Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib: a) Plot a histogram of the data with 30 bins. b) Overlay a line plot representing the normal distribution's probability density function (PDF). c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability. d) Set the title of the plot as 'Histogram with PDF Overlay'.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay a line plot representing the normal distribution's PDF
# Define the range for the x values
x = np.linspace(-4, 4, 1000)  # Range from -4 to 4
pdf = norm.pdf(x, loc=0, scale=1)  # PDF of the normal distribution

plt.plot(x, pdf, 'r', linewidth=2, label='PDF')

# c) Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Add a legend
plt.legend()

# Show the plot
plt.show()
````

#Quetion 17 : Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.



````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)  # Random x values
y = np.random.randn(100)  # Random y values

# Create a DataFrame and assign quadrants based on x and y
data = pd.DataFrame({'X': x, 'Y': y})

# Define a function to determine the quadrant
def determine_quadrant(row):
    if row['X'] > 0 and row['Y'] > 0:
        return 'Quadrant 1'
    elif row['X'] < 0 and row['Y'] > 0:
        return 'Quadrant 2'
    elif row['X'] < 0 and row['Y'] < 0:
        return 'Quadrant 3'
    elif row['X'] > 0 and row['Y'] < 0:
        return 'Quadrant 4'
    else:
        return 'Origin'

# Apply the function to determine the quadrants
data['Quadrant'] = data.apply(determine_quadrant, axis=1)

# Create the scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='viridis', style='Quadrant', markers={'Quadrant 1': 'o', 'Quadrant 2': 's', 'Quadrant 3': '^', 'Quadrant 4': 'D'}, legend='full')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title of the plot
plt.title('Quadrant-wise Scatter Plot')

# Show the legend
plt.legend(title='Quadrants')

# Show the plot
plt.show()
````

#Quetion 18 :With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output to display in a notebook (use output_file() for standalone HTML)
output_notebook()

# Generate x values and compute the sine of those values
x = np.linspace(0, 2 * np.pi, 100)  # 100 points from 0 to 2
y = np.sin(x)  # Sine function

# Create a figure
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y',
           width=800, height=400)

# Add grid lines
p.grid.grid_line_alpha = 0.5

# Add a line renderer
p.line(x, y, legend_label="Sine Wave", line_width=2, line_color="blue")

# Show the plot
show(p)
````

#Quetion 19 :Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.


````python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Prepare the output to display in a notebook
output_notebook()

# Generate random categorical data
categories = [f'Category {i}' for i in range(1, 11)]  # 10 categories
values = np.random.randint(10, 100, size=10)  # Random values for each category

# Generate random colors for each category
colors = ['#'+ ''.join(np.random.choice(list('0123456789ABCDEF'), size=6)) for _ in values]

# Create a DataFrame
data = pd.DataFrame({'Category': categories, 'Value': values, 'Color': colors})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values',
           width=800, height=400)

# Add hover tooltips
hover = HoverTool(tooltips=[("Category", "@Category"), ("Value", "@Value")])
p.add_tools(hover)

# Add bars to the figure with colors from the source
p.vbar(x='Category', top='Value', source=source, width=0.9, fill_color='Color')

# Show the plot
show(p)
````

#Quetion 20 : Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.


````python
import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.random.rand(100)  # Random values

# Create a line plot
fig = go.Figure()

# Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data', line=dict(color='blue')))

# Label the axes
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()

````
